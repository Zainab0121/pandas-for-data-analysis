## SELECTING AND INDEXING
- Use of .loc - The.loc[] function is used for label-based indexing. It allows us to access rows and columns by their labels.
- Use of .iloc - The .iloc[] function is used for position-based indexing. It allows us to access rows and columns by their integer positions. 


```python
# Selecting a single row (Entire row)
df.loc[2] # Note this number is not the position label, it is the index label
```

```python
# Selecting a single row but a specific column
df.loc[2, 'Name'] # Literal translation of this is, go to row named/labeled 2 and give me the Name 
```

```python
# Selecting Multiple rows 
# Always remember to use double brackets for multiple selections
df.loc[[2, 4, 6, 8], 'Name']
```

```python
#Selecting all rows but specific columns
df.loc[: , ['Name', 'Age', 'Cabin']]
```

```python
# Selecting from a specific range of row labels
df.loc[1:5 , ['PassengerId', 'Name']]
```

```python
# Here we are selecting a row based on its positional index
df.iloc[5]
```

```python
# Selecting a single row but a specific column
# In .iloc, both the row and column are represented by their positional index not label
df.iloc[121, 3]
```

```python
# Selecting multiple rows based on their positional index
# Note:- Unlike in label indexing where a range of 0:5 includes all the values in that range including the first and last item, 
# The positional index does not include the last item of a range hence range 0:5 gives you items in index 0 to 4 
df.iloc[0:6]
```

```python
# Select multiple specific rows and one specific column
df.iloc[[1, 3, 5, 7, 9], 3]
```

```python
# Select all rows and specific columns
df.iloc[:, [1, 2, 3]]
```

## SORTING + NEW COLUMNS

 - sort_values() function sorts a DataFrame by one or more columns in ascending or descending order.
   
- Syntax: DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')

Parameter: 

- by: Column name(s) to sort by (single or list).
- axis: 0 or 'index' for sorting rows; 1 or 'columns' for sorting columns.
- ascending: Boolean; True for ascending, False for descending.
- inplace: Boolean; if True, modifies the original DataFrame.
- kind: Sorting algorithm: 'quicksort', 'mergesort', or 'heapsort'.
- na_position: 'first' or 'last'; defines position of NaN values. Default is 'last'.
- The output returns a sorted DataFrame with the same dimensions as the original.

- by → what to sort with
- axis → sort rows or columns
- ascending → direction
- inplace → overwrite or not
- kind → algorithm used
- na_position → where NaNs go


```python
# Sort by a single column 
fare_sort = df.sort_values('Fare', ascending = False)
fare_sort.head(10)
```

```python
# Sorting using multiple columns
df.sort_values(['Age', 'Pclass'])
```

```python
# Sorting using multiple columns, priority based
# This code says, sort by age first, if age is same, sort by class
# In multicolumn sorting, the second column acts as a tie-breaker
df.sort_values(by = ['Age', 'Pclass'], ascending = [False , True])
```

```python
# Handling missing values
missing_cabins = df.sort_values('Cabin', na_position = 'first')
missing_cabins.head(10)
```

## MISSING DATA

Missing data occurs due to many reasons such as lack of data during data entry, joins, etc.


Pandas treats all these as missing data:
- NaN → numeric missing values
- None → object columns
- NaT → missing datetime

#### isnull/isna

What is the difference between .isnull() and .isna()?
- They are aliases — both perform the same function and detect missing values in a pandas object.

There is no functional difference:
-  Both detect missing values
-  .isnull() → SQL-style naming
-  .isna() → NumPy/pandas-style naming
-  Choose one and be consistent

```python
# Checks every cell to tell you which are missing data and which are not
# True indicates missing data, while False indicates that data is present
df.isnull()
```

```python
# This tells you the total number of missing data in each column
df.isnull().sum()
```

```python
# This gives you the percentage of missing data in each column
(df.isnull().sum() / len(df)) * 100
```

```python
# Show me all the rows that have at least one missing value.
df[df.isnull().any(axis=1)] # axis = 1 means data is checked row-wise
```

```python
# Checking a specific column for missing data
df['Cabin'].isnull().sum()
```

###### When to use mean/ median / mode?
- For numeric data with a normal/ symmetric distribution we use mean
- For numeric data that is skewed or has outliers we use median
- For any categorical data we use mode

###### How do you know if your data is skewed?
1. Compare Mean and Median
   - Mean ≈ Median → roughly symmetric
   - Mean > Median → right skewed (long tail to the right)
   - Mean < Median → left skewed (long tail to the left)

   
2. Use the Skew function:- df['column'].skew()
    - Skew = 0 → symmetric
    - Skew > 0 → right skew
    - Skew < 0 → left skew


```python
df['Age'].skew()
```

## GROUPBY OPERATION
Involves three main steps:- 
- Split data into groups
- Apply an operation to each group
- Combine results back together

SYNTAX: - df.groupby("COLUMN_TO_GROUP")["COLUMN_TO_CALCULATE"].OPERATION()

```python
df.groupby('Sex').count()
```

```python
df.groupby('Survived')['Age'].mean()
```

```python
df.info()
```

```python
# Number of total passangers per gender
df.groupby('Sex')['Survived'].count()
```

```python
# Number of survivors per gender
df.groupby('Sex')['Survived'].sum()
```

```python
# In groupby using multiple columns, each unique combination becomes a group
df.groupby(['Pclass', 'Sex'])['Survived'].count()
```

```python
# .reset_index() helps change the output from a Multi-index to regular columns
df.groupby(['Pclass', 'Sex'])['Survived'].count().reset_index()
```

#### AGG, TRANSFORM and APPLY

- agg() -  make a summary table
- transform() -  return result for every original row
- apply() - run custom logic


##### AGG
Means:- Give me one result per group.

Use agg() when:
- making reports
- dashboards
- summary tables
- statistics

```python
# Averge fare paid by each gender
df.groupby("Sex")["Fare"].agg("mean")
```

```python
df.groupby("Sex")["Fare"].agg(["mean", "sum", "count"])
```

```python
# Here we have different calculations per column
# Average fare paid and total survivors all by gender
df.groupby("Sex").agg({
    "Fare": "mean",
    "Survived": "sum"
})
```

##### TRANSFORM
Means:- Calculate something per group but keep all rows.

Use transform() when:
- adding new columns
- normalization
- ML features
- comparing row to group

```python
# Find how much more a passenger paid in comparison to the avearge fare of their respective class
df["Paid_vs_Avg"] = df["Fare"] - df.groupby("Pclass")["Fare"].transform("mean")
df.head()
```

```python
# For every row the mean fare is added based on the gender in that row
df.groupby("Sex")["Fare"].transform("mean")
```

##### APPLY
Means:- Let me run my own function on each group.

Use apply() when:
- special logic needed
- ranking inside groups
- complex filtering

```python
# take top 2 passengers with highest fare inside each gender
df.groupby("Sex").apply(lambda x: x.nlargest(2, "Fare"))
```

```python
# To avoid the above warning, this excludes the grouping columns
df.groupby("Sex").apply(lambda x: x.nlargest(2, "Fare"), include_groups = False)
```
